```python
class Card():

    def __init__(self,suit,number):
        self.suit = suit
        self.number = number
        print("Card Created")
        if number==1:    
            print (f"This card is the Ace of {self.suit}") 
        elif number==11:
            print (f"This card is the Jack of {self.suit}")
        elif number==12:    
            print (f"This card is the Queen of {self.suit}")
        elif number==13:    
            print (f"This card is the King of {self.suit}")
        else:
            print (f"This card is the {self.number} of {self.suit}")
        
    def __str__(self):
        if self.number==1:    
            return (f"The Ace of {self.suit}") 
        elif self.number==11:    
            return (f"The Jack of {self.suit}")
        elif self.number==12:    
            return (f"The Queen of {self.suit}")
        elif self.number==13:    
            return (f"The King of {self.suit}")
        else:
            return (f"The {self.number} of {self.suit}")
    def __del__(self):
        print (f"The {self.number} of {self.suit} has been destroyed")
```

```python
my_deck =[]

def DeckCreator():
    
    for i in range(1,14):
        for j in ['hearts','diamonds','clubs','spades']:
            my_deck.append(Card(j,i))

    return len(my_deck)
```

```python
class Account:
    
    def __init__(self,owner,balance):
        
        self.owner = owner
        self.balance = balance
        
    def __str__(self):
        return (f" Account owner: {self.owner}\n Account balance: {self.balance}")
        
    def deposit(self,deposit_amount):
        self.deposit_amount = deposit_amount        
        self.balance = self.balance + self.deposit_amount
        
        print ("Deposit Accepted")
    
    def withdrawal(self,withdrawal_amount):
        self.withdrawal_amount = withdrawal_amount
        
        if self.withdrawal_amount <= self.balance:
            self.balance = self.balance - self.withdrawal_amount
            print ("Withdrawal accepted")
        else:
            print ("Withdrawal rejected - insufficient funds")
```

```python
def ScoreCalculator(my_deck,a,b):
    user1cardnumber = my_deck[a].number
    user2cardnumber = my_deck[b].number
    if user1cardnumber in [11,12,13]:
        user1cardnumber = 10
    if user2cardnumber in [11,12,13]:
        user2cardnumber = 10
    return user1cardnumber + user2cardnumber
```

```python
def WinCheck(user_score,dealer_score,user_initial_bet):
    if dealer_score>=22:
        print("DEALER - BUST")
        print("YOU WIN!")
        user_account.balance = user_account.balance +(2*user_initial_bet)
    elif dealer_score>user_score:
        print ("DEALER WINS")
    elif dealer_score<user_score:
        print ("YOU WIN")
        user_account.balance = user_account.balance +(2*user_initial_bet)
    elif dealer_score==user_score:
        print("Tie game")
        user_account.balance = user_account.balance + user_initial_bet
```
